```python
import json
import arrow

file_path = '/Users/ianscott/Development/kcr-untracked-files/records-for-import.json'

with open(file_path) as file:
    data = json.load(file)

    first_object = iter(d for d in data if d["embargoed"] != "no" and d["embargo_end_date"] and arrow.get(d["embargo_end_date"], 'MM/DD/YYYY') > arrow.utcnow()).__next__()
    pretty_printed_object = json.dumps(first_object, indent=4)

    print(pretty_printed_object)
```
